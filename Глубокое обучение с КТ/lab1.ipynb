```python
class RBFTrain:
    def __init__(self, input_dim, output_dim, sigma=1.0):
        self.W = np.random.randn(output_dim, input_dim) * np.sqrt(2 / input_dim)
        self.b = np.zeros((output_dim, 1))
        self.sigma = sigma

    def forward(self, x):
        self.x = x
        distances = np.sum((x[:, np.newaxis, :] - self.W[np.newaxis, :, :]) ** 2, axis=2)
        kernel = np.exp(-distances / (2 * self.sigma ** 2))
        self.out = kernel + self.b.T
        return self.out

    def backward(self, grad_output, grads, layer_index):
        grad_z = grad_output * self.activate_derivative(self.y)
        grad_W = np.dot(grad_z, self.x.T)
        grad_b = np.sum(grad_z, axis=1, keepdims=True)
        grad_x = np.dot(self.W.T, grad_z)
        return grad_x
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score

def train_models(X_train, y_train, X_test, y_test, layer_configs, epochs=50, lr=0.01):
    print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    results = {}

    for num_layers in layer_configs:
        layers = [LinearTrain(X_train.shape[0], 10, activation='relu')]
        for _ in range(num_layers - 1):
            layers.append(LinearTrain(10, 10, activation='relu'))
        #layers.append(LinearTrain(10, y_train.shape[0], activation='identity'))
        layers.append(LinearTrain(10, y_train.shape[0] if len(y_train.shape) > 1 else 1, activation='identity'))

        optimizer = Adagrad(lr=lr)
        loss_history = []

        for epoch in range(epochs):
            total_loss = 0

            #  
            x = X_train
            for layer in layers:
                x = layer.linear_transform(x)

            loss, grad = soft_argmax_cross_entropy(x, y_train)
            total_loss += loss

            #  
            grads = {}
            grad_output = grad
            for i in range(len(layers) - 1, -1, -1):
                grad_output = layers[i].backward(grad_output, grads, i)


            for i, layer in enumerate(layers):
                layer.W, layer.b = optimizer.update({'W': layer.W, 'b': layer.b}, {'W': grads[f'W{i}'], 'b': grads[f'b{i}']}).values()

            loss_history.append(total_loss)

        x = X_test
        for layer in layers:
            x = layer.linear_transform(x)
        preds = np.argmax(x, axis=0)
        true_labels = np.argmax(y_test, axis=0)

        if isinstance(true_labels, np.int64):
            true_labels = np.array([true_labels])
        if isinstance(preds, np.int64):
            preds = np.array([preds])

        acc = accuracy_score(true_labels, preds)
        f1 = f1_score(true_labels, preds, average='weighted')

        results[num_layers] = {'loss_history': loss_history, 'accuracy': acc, 'f1': f1}

    plt.figure(figsize=(10, 6))
    for num_layers, data in results.items():
        plt.plot(data['loss_history'], label=f'{num_layers} ')
    plt.xlabel('')
    plt.ylabel('')
    plt.legend()
    plt.title(' ')
    plt.show()

    return results

class LinearTrain:
    def __init__(self, input_dim, output_dim, activation='identity'):
        self.W = np.random.randn(output_dim, input_dim) * 0.01
        self.b = np.zeros((output_dim, 1))
        self.activation = activation

    def activate(self, x):
        if self.activation == 'identity':
            return x
        elif self.activation == 'relu':
            return np.maximum(0, x)
        elif self.activation == 'tanh':
            return np.tanh(x)

    def activate_derivative(self, x):
        if self.activation == 'identity':
            return np.ones_like(x)
        elif self.activation == 'relu':
            return (x > 0).astype(float)
        elif self.activation == 'tanh':
            return 1 - np.tanh(x) ** 2

    def linear_transform(self, x):
        self.x = x
        self.y = np.dot(self.W, x) + self.b.reshape(-1,1)
        return self.activate(self.y)

    def backward(self, grad_output, grads, layer_index):
        grad_z = grad_output * self.activate_derivative(self.y)
        grad_W = np.dot(grad_z, self.x.T)
        grad_b = np.sum(grad_z, axis=1, keepdims=True)
        grad_x = np.dot(self.W.T, grad_z)

        grads[f'W{layer_index}'] = grad_W
        grads[f'b{layer_index}'] = grad_b

        return grad_x

class Adagrad:
    def __init__(self, lr=0.01, eps=1e-8):
        self.lr = lr
        self.eps = eps
        self.G = {}

    def update(self, params, grads):
        updated_params = {}
        for step, value in params.items():
            if step not in self.G:
                self.G[step] = np.zeros_like(value)
            self.G[step] += grads[step] ** 2
            updated_params[step] = value - self.lr * grads[step] / (np.sqrt(self.G[step]) + self.eps)
        return updated_params

def soft_argmax_cross_entropy(x, y):
    exp_x = np.exp(x - np.max(x, axis=0, keepdims=True))
    p = exp_x / np.sum(exp_x, axis=0, keepdims=True)
    grad_x = (p - y)
    return -np.sum(y * np.log(p + 1e-9)) / y.shape[0], grad_x
```
